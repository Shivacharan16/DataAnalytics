# Dictionary data structure

## creating differenty type of dictionary

```python
# iteam structure in dictionary : (key : value) pair
e_d = {}

emp_d = dict()
print(emp_d)
print(type(e_d))
```

```python
#{key1:value1,key2:value2,.,.,}
vt_int_k_d = {1:"bendakaya",2:"doskaya", 3 :"sorakaya",4:"vankaya"}
print(vt_int_k_d)
```

```python
vt_int_k_d = {1.1:"bendakaya",2.1:"doskaya", 3.1:"sorakaya",4.1:"vankaya"}
print(vt_int_k_d)
```

```python
s = {"bendakaya":40,"doskaya":80,"sorakaya":40,"vankaya":25}
print(s)
```

```python
s1 = {1:"shiva",2.8:"charan","bikkumalla":25}
print(s1)
```

```python
# can u use tuple as a key
t_k_d ={(1,2,3):"shiva",(40,80):"bendakaya"}
print(t_k_d)
```

```python
# we cannot list as key in dist
l_k_d ={[1,2,3]:"shiva",[40,80]:"bendakaya"}
print(l_k_d)
```

```python

t_k_d ={{1,2,3}:"shiva",{40,80}:"bendakaya"}
print(t_k_d)
```

```python
# nested dictionary
n_d = {"vt":{"doskaya": 60,"bendaakaya":40}}
len(n_d)
```

```python
n_d["vt"]["bendaakaya"]
```

```python
# dictionary with different type of values 
v_d = {1:10,2:20.5,3:"shiva",4:(20,40),5:[20,50],6:{25,35},7:True,8:False}
print(v_d)
       
```

```python
# we can use list & set as the value
# we cannot use the list & set as the key

# keys must be immutable,but values can be mutable or immutable
```
