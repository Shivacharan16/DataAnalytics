# Accessing the dictionary items

```python
# we can access the values in a item through its key

hw_h = {"tom cruise":1000,"angellina":800,"tony":300}

val =hw_h["angellina"]
print(val)
```

```python
val2 =hw_h["pawan kalyan"]
print(val2)
```

```python
# get function
val3 =hw_h.get("angellina")
print(val3)
```

```python
val4 =hw_h.get("pawan kalyan")
print(val4)
```

```python
if "angellina" in hw_h:
    print("Hurray,we find angellina")
else:
    print("we miss angellina")
    
```

```python
if "pawan kalyan" in hw_h:
    print("Hurray,we find pawan kalyan")
else:
    print("we miss pawan kalyan")
```

```python
for k in hw_h:
    print(k)
```

```python
for k in hw_h:
    print(k,":",hw_h[k])
```

```python
# keys(): return all the keys of the dictionary
all_keys = hw_h.keys()
print(all_keys)
```

```python
# values(): retuen all the values
all_values = hw_h.values()
print(all_values)
```

```python
for k in hw_h.values():
    print(k)
```

```python
# item(): function returns all the items in the form of list of tuples 
hw_h.items()
```

# inserting or updating dictionary items

```python
s_x = {"name":"angelina",'age':70}

s_x["age"]=16  #update
print(s_x)
```

```python
s_x["city"]="kukatpalli" # insert
print(s_x)
```

```python
# update(): insert or update multiple items together

s_x.update(city = "pista house 3rd floor", age = 18)
print(s_x)
```

```python
s_x.update(mobile = 101,salary = 1000, height = 6.2, weight = 40)
print(s_x)
```

```python
## delete  items from dictionary (pop(),popitem(),clear(),del
```

```python
# pop(): remove the item with the specified key
s_x.pop("age")
```

```python
print(s_x)
```

```python
# popitem(): delete last inserted item
s_x.popitem()
```

```python
print(s_x)
```

```python
# clear(): delete all the items form the dictionary
s_x.clear()
s_x
```

```python
del s_x
```
