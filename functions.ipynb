## default parameters

```python
# Example-1:
def fun(p1 = 'a'): 
    print(p1)

fun('c')
```

```python
# Example-2 :
def fun(p1 = 'a',p2 = 'b'):
    print(p1,p2)
fun('z','x')
```

```python
# Example-3 : always define the default parameters from right side
def fun(p1,p2 ='a'):
    print(p1,p2)
fun('z')
```

## Passing arbitrary number of arguments using *arg

```python
# Q: write a function which receives  5 values and print thouse values
def display(*hello): # we can use any variable after '*' in the place of arg
    for i in hello:
        print(i)
display()
display(28,55)
display(12,5,16,47,57)
```

## passing key-value arbitrary arguments using **karg

```python
def shiva(**charan):
    for n in charan:
        print(n,charan[n])        
shiva(sai = 20,cherry = 30)
```

# lambda function

```python
# Example-1: structure of the lambda function
# syntax : obj = lambda arguments :expression
# addibng 10 inside the input value
# here a is the input argument, a+10is expression which will be elevluated and return the final value
obj = lambda a: a+10

print(obj(10))
```

```python
obj = lambda a,b:a+b+10

print(obj(10,20))
```

```python
# Example-2: lambda function inside the other function
def shiva(power):
    obj = lambda a:a**power
    return obj
obj1 = shiva(2)
obj2 = shiva(3)
print(obj1(10))  # this is lambda function object
print(obj2(5))
```

```python
obj3 = shiva(5)
print(obj3(5))
```

```python
# interview questions:
# sort this list of tuples based upon the study hours

students = [('shiva',10),('harish',8),('niranjan',6),('cherry',12)]

# here lambda function is telling to the sorted function to use 1th value for sorting
sorted(students,key = lambda x:x[1])

```

```python
students = [('shiva',10,85),('harish',8,75),('niranjan',6,70),('cherry',12,80)]
sorted(students,key = lambda x:x[2])
```
