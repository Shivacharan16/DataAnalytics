# Match case statement

```python
# Match case statement is used to replace multiple if-else conditions.
# Match case is similaer to switch case in c & c++.
# Match case more advance then the switch case.
# Match case is haveing advance features like handlingof data structures and conditions.
```

```python
# Example-1: structure of match case
num = int(input("enter a number"))

match num:
    case 10:
        print("shiva")
    case 20:
        print("charan")
    case _:
        print("something else")
```

```python
# Example-2: match case with data structure
data = [10,20,30]

match data:
    case [1,2,3]:
        print("Hello 1")
    case [30,20,10]:
        print("Hello 2")
    case [10,20,30]:
        print("Hello 3")
    case _:
        print("Nothing match")
```

```python
# Example-3: Match caase  using condition
num = int(input("enter a num"))

match num:
    case x if x>0:
        print("It is postive num")
    case y if y <0:
        print("It is negative num")
    case _:
        print("It is zero")
```

## Insert iteam in set(add(), update())


```python
set1 = {10,20.5,"shiva"}

# add(): it is used to insert single iteam in the set
set1.add("charan")
print(set1)

set1.add(2004)
print(set1)

# update(): this method is used to insert multiple items together inside the set

items = ["samaantha","kajal","pooja","shruthi"]
set1.update(items)
print(set1)
```

## delete items from the set (remove(),dicard(),pop(),clear(),del())

```python
d_l = {"campa","mazza","sprite","tumbsup","sting","sugarcane","coccnut water"}

#remove() : delete particular one item whose name is provided
# it will show the erroe if item don't exist
d_l.remove("sprite")

print(d_l)

# dicard :delete particular one item whose name is provided
# it will not show any erroe if item don't exist
d_l.discard("tumbsup")
print(d_l)
```

```python
# pop() : this will delete Random item
d_l.pop()
print(d_l)
```

```python
# clear(): delete all the elements
d_l.clear()
print(d_l)
```

```python
del d_l
print(d_l)
```

```python
# containment operation : check if element exist in the set or not

s_f = {"pongal","Dussehra","bonalu","onam","ugadi"}

if "Holi" not in s_f:
    print("happy holi")
else:
    print("come to North India")
```

```python
#  set comparison  operator
set1 = {1,2,3,4}
set2 = {4,2,3,1}
if set1==set2:
    print('match')
else:
    print('not match')
if set1!=set2:
    print('match')
else:
    print('not match')
```
